Just a basic CSV data cleaning project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Starfinder Alien Archives - Sheet1.csv')
df.shape
df.CR.value_counts
df['CR'] = df['CR'].str.lower()
df['CR'] = df['CR'].str.strip()
#df['CR'] = np.where(.5, .5, df.CR)
#df['CR'] = np.where(1/3, .33, df.CR)
#df["CR"] = df["CR"].astype(int)

tf = df.CR.unique()
tf
```

```python
CR_count = []
pie_angles =[]
explode_values = []


def get_total(inlist, outlist):
    for i in inlist:
        outlist.append((df['CR'] == i).sum())

get_total(tf,CR_count)
def get_angles(inlist):
    outlist = []
    for i in inlist:
        outlist.append((i/sum(inlist))*360)
    return outlist


def get_explode(inlist):
    explode_values = []
    for i in inlist:
        explode_values.append(.1)
    return explode_values
        
pie_angles = get_angles(CR_count)
explode_values = get_explode(CR_count)
CR_count
len(explode_values)
```

```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = tf
sizes = pie_angles
explode = (explode_values)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = tf,explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(tf,loc = 'best')
plt.tight_layout()
plt.show()
```
